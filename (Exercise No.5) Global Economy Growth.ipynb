# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Reading the data

```python
df = pd.read_csv("Global Economy Growth.csv")
df
```

```python
df.info()
```

#  Data cleaning and processing
Remove '%' sign and convert to float

```python

df['Actual (2021)'] = df['Actual (2021)'].str.rstrip('%').astype('float')
df['Actual (2022)'] = df['Actual (2022)'].str.rstrip('%').astype('float') 
df['Estimates (2023)'] = df['Estimates (2023)'].str.rstrip('%').astype('float') 
df['Estimates (2024)'] = df['Estimates (2024)'].str.rstrip('%').astype('float') 

# Display the updated DataFrame
print(df)
```

# Data Visualization
using (lineplot, Barplot)

```python
growth_rates= ["Global Economy", "Advanced Economies", "Emerging Market & Developing Economies",
"USA", "China", "Japan", "India", "Euro Area", "Saudi Arabia"]
# Plotting growth rates
plt.figure(figsize=(14, 7))
for Growth in growth_rates:
    plt.plot(df.columns[1:], df[df["Growth rates"] == Growth].values[0][1:], label=Growth)

plt.title("Growth Rates")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_melted = df.melt(id_vars=["Growth rates"], 
                              var_name="Year", 
                              value_name="Growth Rates (%)")

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Year", y="Growth Rates (%)", hue="Growth rates")
plt.title('Growth Rates (2021-2024)')
plt.ylabel('Growth Rates (%)')
plt.xlabel('Year')
plt.legend(title='Growth RatesType')
plt.grid(True)
```

# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Reading the data

```python
df = pd.read_csv("Sectoral Expenditure.csv")
df

```

```python
df.info()
```

# Data cleaning and processing
Remove '%' sign and convert to float

```python
# Remove '%' sign and convert to float
df['AnnualChange (Budget 2024 -Estimates 2023)'] = df['AnnualChange (Budget 2024 -Estimates 2023)'].str.rstrip('%').astype('float') 

# Display the updated DataFrame
print(df)
```

# Data Visualization
using (lineplot, Barplot)

```python
plt.figure(figsize=(10, 6))
plt.barh(df['Sector'], df['AnnualChange (Budget 2024 -Estimates 2023)'], color='skyblue')
plt.xlabel('AnnualChange (Budget 2024 -Estimates 2023)')
plt.title('Annual Percentage Change by Sector(Budget 2024 -Estimates 2023)')
plt.grid(axis='x')
```

```python
plt.figure(figsize=(10, 6))
plt.barh(df['Sector'], df['Budget (2024)'], color='skyblue')
plt.xlabel('Budget (2024)')
plt.title('Budget (2024) by Sector')
plt.grid(axis='x', color='g', alpha=0.35,linestyle=':')
```

```python
df_melted = df.melt(id_vars=["Sector"], 
                              var_name="Year", 
                              value_name="Sector Rate (%)")

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Year", y="Sector Rate (%)", hue="Sector")
plt.title('Sector Rates (2022-2024)')
plt.ylabel('SectorRate (%)')
plt.xlabel('Year')
plt.legend(title='Sector Type')
plt.grid(True)
```

```python
df_melted = df.melt(id_vars=["Sector"], 
                              var_name="Year", 
                              value_name="Sector Rate (%)")

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x="Year", y="Sector Rate (%)", hue="Sector")
plt.title('Sector Rates (2022-2024)')
plt.ylabel('SectorRate (%)')
plt.xlabel('Year')
plt.legend(title='Sector Type')
plt.grid(True)
```

```python
plt.figure(figsize=(15, 5))
sns.barplot(x='Sector', y='Budget (2024)', data=df, palette='hls')
plt.title('Sector Vs Budget (2024)', fontsize=15)
plt.xlabel('Sector', fontsize=15)
plt.ylabel('Average Budget (2024)', fontsize=15)
plt.xticks(rotation=45)  
plt.show()
```

```python
Sec_Bud= df.groupby('Sector')[['Actual (2022)','Estimates (2023)','Budget (2024)']].sum()
Sec_Bud
```

```python
Sec_Bud.plot(subplots=True)
plt.xticks(rotation=85)  
plt.show()
```

```python
plt.figure(figsize=(50, 50))
Sec_Bud.plot()
plt.xticks(rotation=85)  
plt.show()
```

# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Reading the data

```python
df = pd.read_csv("Inflation.csv")
df
```

```python
df.info()
```

# Data cleaning and processing
Remove '%' sign and convert to float

```python
# Remove '%' sign and convert to float
df['Actual (2021)'] = df['Actual (2021)'].str.rstrip('%').astype('float')
df['Actual (2022)'] = df['Actual (2022)'].str.rstrip('%').astype('float') 
df['Estimates (2023)'] = df['Estimates (2023)'].str.rstrip('%').astype('float') 
df['Estimates (2024)'] = df['Estimates (2024)'].str.rstrip('%').astype('float') 

# Display the updated DataFrame
print(df)
```

# Data Visualization
using (lineplot, Barplot)

```python
df_melted = df.melt(id_vars=["Inflation"], 
                              var_name="Year", 
                              value_name="Inflation Rate (%)")

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Year", y="Inflation Rate (%)", hue="Inflation")
plt.title('Inflation Rates (2021-2024)')
plt.ylabel('Inflation Rate (%)')
plt.xlabel('Year')
plt.legend(title='Inflation Type')
plt.grid(True)
```

```python
df_melted = df.melt(id_vars=["Inflation"], 
                              var_name="Year", 
                              value_name="Inflation Rate (%)")

# Create a line plot using Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_melted, x="Year", y="Inflation Rate (%)", hue="Inflation", marker='o')
plt.title('Inflation Rates (2021-2024)')
plt.ylabel('Inflation Rate (%)')
plt.xlabel('Year')
plt.legend(title='Inflation Type')
plt.grid(True)

# Show the plot
plt.show()
```
